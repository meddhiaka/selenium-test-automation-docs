# Login automation with selenium & getting some data of the website...

```python
from selenium import webdriver
import os
from selenium.webdriver.common.by import By
```

> Getting user and password credentials from local python environment

```python
user = os.getenv("EXAMPLE_USER")
password = os.getenv("EXAMPLE_PASSWORD")
```

```python
url = "https://www.stealmylogin.com/demo.html"
```

```python
driver = webdriver.Chrome()
```

```python
driver.get(url)
```

```python
driver.find_element(By.NAME, "username").send_keys(user)
driver.find_element(By.NAME, "password").send_keys(password)
```

> Now the credentails are injected successfully in the form:
> 


```python
driver.find_element(By.CSS_SELECTOR, 'input[type="submit" i]').click()
```

> the login button is click and we are redirected to another website
